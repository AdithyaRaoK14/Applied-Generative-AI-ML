# **Q3**

**(a) Apply logistic regression algorithm for the given dataset.**

**(b) Analyse the result of the logistic regression algorithm for the given dataset.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data=pd.read_csv('/content/drive/MyDrive/AGAI&ML Lab Midsem Dataset/iris 2.csv', header=None)
data = data[0].str.split(',', expand=True)
```

```python
data
```

```python
x=data.iloc[:,:4].values
x
```

```python
y=data.iloc[:,4].values
y
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)
y_encoded
```

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y_encoded,test_size=0.35,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
#standard scalar
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(random_state=0)
model.fit(x_train_scaled,y_train)
```

```python
y_pred=model.predict(x_test_scaled)
```

```python

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

**use built-in function in Numpy and Pandas to find the dot product of a=((0,1)(1,1)) and b= ((4,1)(2,1)).**

```python
import numpy as np

A = np.array([[0, 1], [1, 1]])
B = np.array([[4, 1], [2, 1]])

dot_product_numpy = np.dot(A, B)
print("Dot product using NumPy:")
print(dot_product_numpy)
```
